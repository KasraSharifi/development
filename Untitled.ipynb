```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
student_performance_raw_dataset = pd.read_csv("Student_Performance.csv")

#Taking care of the categorical column we have
student_performance_raw_dataset.loc[
    student_performance_raw_dataset["Extracurricular Activities"] == "No", 
    "Extracurricular Activities"] = 0

student_performance_raw_dataset.loc[
    student_performance_raw_dataset["Extracurricular Activities"] == "Yes", 
    "Extracurricular Activities"] = 1

student_performance_raw_dataset["Extracurricular Activities"] = (
    student_performance_raw_dataset["Extracurricular Activities"].apply(lambda x:int(x))
    
)

student_performance_raw_dataset.sample(1000).describe()
```

```python
feature_space = (student_performance_raw_dataset[student_performance_raw_dataset.columns[:5]]
                                    .values)
target_value = student_performance_raw_dataset["Performance Index"]

feature_space_torch = torch.tensor(feature_space)
target_value = torch.tensor(target_value).float()
target_value = target_value.reshape(feature_space_torch.shape[0],1)

#Normalize the data by using max scaler
feature_space_normalized_torch = feature_space_torch / feature_space_torch.max(dim = 0)[0]

# Create our train-test-split. Here, because we wan to develop only one model, 
# we are not gonna define dev set
(feature_space_normalized_torch_train, feature_space_normalized_torch_test, target_values_train,
target_values_test) = train_test_split(feature_space_normalized_torch, target_value,
                                      train_size=0.9)


#Create our torch.dataset
train_tensor_dataset = TensorDataset(feature_space_normalized_torch_train, target_values_train) 
test_tensor_dataset = TensorDataset(feature_space_normalized_torch_test, target_values_test) 


#Create our DataLoader
train_data_loader = DataLoader(train_tensor_dataset, batch_size = 64, drop_last = True)
test_data_loader = DataLoader(test_tensor_dataset, batch_size = len(test_tensor_dataset))
```

```python
class AnnModel(nn.Module):
    def __init__(self):
        super().__init__()
    
        self.layers_dictionary = nn.ModuleDict()
        self.layers_dictionary["INPUT_LAYER"] = nn.Linear(5, 16)
        self.layers_dictionary["HIDDEN_LAYER_1"] = nn.Linear(16, 16)
        self.layers_dictionary["OUTPUT_LAYER"] = nn.Linear(16, 1)
    
    def forward(self, feature_space):
        
        # If we don't use any activation function here, our model reduces to be a simple 
        # linear regression
        input_layer_output = F.relu(self.layers_dictionary["INPUT_LAYER"]
                                              (feature_space))
        hidden_layer_output = F.relu(self.layers_dictionary["HIDDEN_LAYER_1"]
                                     (input_layer_output))
        output_layer_output = self.layers_dictionary["OUTPUT_LAYER"](hidden_layer_output)
        
        return output_layer_output
```

```python
#Test our AnnModel class with a simple example

test_model = AnnModel()
test_model(next(iter(train_data_loader))[0][0])
```

```python
def train_ann_model(ann_model_class, training_data_loader, testing_data_loader, epoch = 500, 
                    learning_rate = 0.01):
    trained_model = ann_model_class
    
    loss_function = nn.MSELoss()
    optimizer_function = torch.optim.SGD(trained_model.parameters(), lr = learning_rate)
    
    loss_values_list = []
    training_mse_list = []
    testing_mse_list = []
    
    for iteration in range(epoch):
        batches_loss_values_list = []
        batch_mse_values_list = []
        
        for training_feature_space_batch, training_taget_value_batch in training_data_loader:
            prediction = trained_model(training_feature_space_batch)
            
            batch_loss_value = loss_function(prediction, training_taget_value_batch)
            
            #Withouth this block of code, we never make any progress, meaning optimization
            # is implemeneted in this stage.
            optimizer_function.zero_grad()
            batch_loss_value.backward()
            optimizer_function.step()
            
            batches_loss_values_list.append(batch_loss_value.detach().item())
            
            batch_mse_values_list.append(torch.mean(torch.pow(torch.abs(prediction - training_taget_value_batch), 2))
                  .detach().item())
            
        trained_model.eval()
        with torch.no_grad():
            test_feature_space, test_taget_value = next(iter(testing_data_loader))
            test_prediction = trained_model(test_feature_space)
            
            testing_mse_list.append(torch.mean(torch.pow(torch.abs(test_prediction - test_taget_value), 2))
                  .detach().item())
           
        loss_values_list.append(np.array(batches_loss_values_list).mean())
        training_mse_list.append(np.array(batch_mse_values_list).mean())
        
    return loss_values_list, training_mse_list, testing_mse_list, trained_model
```

```python
training_loss, training_mse_list, testing_mse_list, trained_model = train_ann_model(AnnModel(), 
                     training_data_loader= train_data_loader, 
                     testing_data_loader = test_data_loader ,epoch = 30, 
                     learning_rate = 0.0001)
```

```python
pd.Series(training_loss[:15]).plot(kind = "line");
```

```python
pd.Series(testing_mse_list, name = "TEST_MSE").plot(legend = True);
pd.Series(training_mse_list, name = "TRAIN_MSE").plot(legend = True)
```
